```python
#  Step 1: Install Required Tools
!apt-get update
!apt-get install -y sra-toolkit bwa samtools bcftools fastqc
```

```python
# Step 2: Download FASTQ Data for SRR9200893
!prefetch SRR9200893
!fastq-dump --split-files SRR9200893
```

```python
# Step 3: Run FastQC
!fastqc SRR9200893_1.fastq -o /content/
!fastqc SRR9200893_2.fastq -o /content/
```

```python
# Step 4: Download and Index Human Reference Genome (GRCh38)
!wget ftp://ftp.ebi.ac.uk/pub/databases/gencode/Gencode_human/release_44/GRCh38.primary_assembly.genome.fa.gz
!gunzip GRCh38.primary_assembly.genome.fa.gz
!bwa index GRCh38.primary_assembly.genome.fa
```

```python
# Step 5: Align Reads to Genome
!bwa mem GRCh38.primary_assembly.genome.fa SRR9200893_1.fastq SRR9200893_2.fastq > aligned.sam
```

```python
#  Step 6: Convert SAM to BAM, Sort and Index
!samtools view -Sb aligned.sam > aligned.bam
!samtools sort aligned.bam -o sorted.bam
!samtools index sorted.bam
```

```python
#  Step 7: Variant Calling
!samtools mpileup -uf GRCh38.primary_assembly.genome.fa sorted.bam | bcftools call -mv -Ov -o variants.vcf
```

```python
# Step 8: Preview First 20 Lines of VCF File
with open('variants.vcf') as f:
    for _ in range(20):
        print(f.readline().strip())
```
